## ggplot
     ggplot(data = dat, mapping = aes())
     
dat tells us what dataframe or tibble to use, and aes() is a general description of the axes. From there, we use + to add additional layers to this plot. Data layers are called geometries and are indicated by the title geom.

```r
print(data(mtcars))
```

```r
data(mtcars)
```

```r
head(mtcars)
```

```r
# To create a histogram, we start with the basic ggplot() function and then add geom_histogram()
ggplot(data = mtcars, aes(x = mpg)) +  geom_histogram()
```

```r
# Depict a grouping variable within each bar in this histogram, by using a fill variable
# am for automatic or manual transmission:

ggplot(data = mtcars, aes(x = mpg, fill = as.factor(am))) +
  geom_histogram()
```

```r
# make a density plot to see the distribution of car weights wt, we could use geom_density()
ggplot(data = mtcars, aes(x = wt)) +
  geom_density()
```

```r
# Create two overlapping density plots grouped by another variable
ggplot(data = mtcars, aes(x = wt, fill = as.factor(vs))) +
  geom_density(alpha = 0.5)
```

## Scatterplots and lines

```r
# look at the relationship between miles per gallon mpg and horsepower hp
ggplot(data = mtcars, aes(x = mpg, y = hp)) +
  geom_point()
```

```r
# Let’s say we wanted to color our data points based on the number of cylinders cyl
ggplot(data = mtcars, aes(x = mpg, y = hp, 
                          color = as.factor(cyl))) +
  geom_point()
```

```r
# connect our datapoints across the x-axis, we could add geom_line() as another layer in our plot
ggplot(data = mtcars, aes(x = mpg, y = hp)) +
  geom_point() +
  geom_line()
```

```r
# Add a more smoothed regression line, we could use geom_smooth()
ggplot(data = mtcars, aes(x = mpg, y = hp)) +
  geom_point() +
  geom_smooth()
```

## Boxplots

```r
# Evaluate the quarter-mile time qsec with the number of cylinders
ggplot(data = mtcars, aes(x = as.factor(cyl), y = qsec)) +
  geom_boxplot()
```

```r
# change the shape, color, and size of our outliers
ggplot(data = mtcars, aes(x = as.factor(cyl), y = qsec)) +
  geom_boxplot(outlier.colour="red", outlier.shape=8,
               outlier.size=4)
```

## Last Thoughts

* The x and y labels as well as the title of a plot can be adjusted by adding
+ labs(x="X axis", y = "Y axis", title = "Title")
* The legend automatically generates if color is specified in the aes() statement. You can avoid generating the legend by specifying color (when relevant) in the geom_ statements.
* If you want to use multiple datasets in one plot, each dataset needs to be associated with the geom_ statements. Call ggplot() without a dataset specified and then specify the dataset in the geom_ statements. However, this may limit the scope of the legend.
* The tidyverse provides extremely thorough and useful cheatsheets on many topics including ggplot().

```r
# add x and y labels and title name
ggplot(data = mtcars, aes(x = as.factor(cyl), y = qsec)) + labs(x="X axis", y = "Y axis", title = "Title") +
  geom_boxplot(outlier.colour="red", outlier.shape=8,
               outlier.size=4)
```
